```python
daily_df['date'] = pd.to_datetime(daily_df['date'])
```

```python
before = daily_df[daily_df['date'] < '2024-04-01']

during = daily_df[
    (daily_df['date'] >= '2024-04-01') &
    (daily_df['date'] <= '2024-07-31')
]
```

```python
print("Before challenge:", before['distance_m'].mean())
print("During challenge:", during['distance_m'].mean())
```

```python
comparison = {
    'Before 75 Hard': before['distance_m'].mean(),
    'During 75 Hard': during['distance_m'].mean()
}

import pandas as pd

comparison_series = pd.Series(comparison)

comparison_series.plot(kind='bar')
plt.title('Activity Before vs During 75 Hard')
plt.ylabel('Average Distance (meters)')
plt.show()
```
